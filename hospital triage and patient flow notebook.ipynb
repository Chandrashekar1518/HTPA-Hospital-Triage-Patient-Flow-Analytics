```python
df=df.drop(columns=["Time of Day"])
```

```python
#time of day
df["Time"]=pd.to_datetime(df["Time"],errors='coerce').dt.time
```

The original time-of-day categorization was inconsistent with the raw time data, so rebuilt it directly from the time column using defined clinical shift boundaries

```python
from datetime import time
def assign_time_of_day(t):
    if time(0,0)<=t<=time(6,0):
        return "Early Morning"
    elif time(6,1)<=t<=time(11,59):
        return "Morning"
    elif time(12,0)<=t<=time(16,59):
        return "Afternoon"
    else:
        return "Evening/night"
df["Time of Day"]=df["Time"].apply(assign_time_of_day)
```

```python
df["Reg_wait%"]=df["Time to Registration (min)"]/df["Total Wait Time (min)"]
df["triage_wait_%"]=df["Time to Triage (min)"]/df["Total Wait Time (min)"]
df["provider_Wait_%"]=df["Time to Medical Professional (min)"]/df["Total Wait Time (min)"]
df.head()
```
